```python
#index reestablished because MySQL couldn't interpret the 'state_dist' column' as the primary key
all_states_df_clean = all_states_df.reset_index()
all_states_df_clean = all_states_df_clean.rename_axis(None)
all_states_df_clean= all_states_df_clean.rename_axis(None, axis='columns')
all_states_df_clean = all_states_df_clean.rename(columns={"index":"state_dist"})
all_states_df_clean.head()
```
