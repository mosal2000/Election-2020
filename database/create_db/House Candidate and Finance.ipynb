```python
import pandas as pd
from sqlalchemy import create_engine

import pymysql
pymysql.install_as_MySQLdb()
```

```python
candidates_file = "Resources/midterm_election_candidates_2018_0921.csv"
candidates_df = pd.read_csv(candidates_file)
candidates_df.head()
```

### House Candidates

```python
House_candidates = candidates_df[candidates_df["branch"]=='H'].copy()
House_candidates.head()
```

```python
House_candidates_1 = House_candidates.drop(['branch'], axis=1)
House_candidates_1.head()
```

```python
connection_string = "root:AbuZaidSZK007@localhost:3306/election_prediction_db"
engine = create_engine(f'mysql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
# Clean the data by dropping duplicates and setting the index
House_candidates_1.drop_duplicates("fec_candidate_id", inplace=True)
```

```python
House_candidates_1.set_index("fec_candidate_id", inplace=True)
```

### Create database connection

```python
House_candidates_1.to_sql(name='house_candidate', con=engine, if_exists='append', index=True)
```

### Candidates Financials

```python
cand_financials_file = "Resources/candidate_summary_2018.csv"
cand_financials_df = pd.read_csv(cand_financials_file)
cand_financials_df.head()
```

```python
#create new table with selected values from cand_financials table. Then, filter it by House and Senate 

cand_financials_cols = [
 'Cand_Id',
 'Cand_Office',
 'Total_Receipt',
 'Total_Disbursement',
 'Cash_On_Hand_COP',
 'Debt_Owed_By_Committee',
 'Coverage_End_Date',
 'Individual_Itemized_Contribution',
 'Individual_Unitemized_Contribution',
 'Individual_Contribution',
 'Other_Committee_Contribution',
 'Party_Committee_Contribution',
 'Cand_Contribution',
 'Total_Contribution',
 'Transfer_From_Other_Auth_Committee',
 'Cand_Loan',
 'Other_Loan',
 'Total_Loan',
 'Offsets_To_Operating_Expenditure',
 'Offsets_To_Fundraising',
 'Offsets_To_Leagal_Accounting',
 'Other_Receipts',
 'Operating_Expenditure',
 'Exempt_Legal_Accounting_Disbursement',
 'Fundraising_Disbursement',
 'Transfer_To_Other_Auth_Committee',
 'Cand_Loan_Repayment',
 'Other_Loan_Repayment',
 'Total_Loan_Repayment',
 'Individual_Refund',
 'Party_Committee_Refund',
 'Other_Committee_Refund',
 'Total_Contribution_Refund',
 'Other_Disbursements',
 'Net_Contribution',
 'Net_Operating_Expenditure',
 'Cash_On_Hand_BOP',
 'Debt_Owe_To_Committee',
 'Coverage_Start_Date']

cand_financials_2 = cand_financials_df[cand_financials_cols].copy()
cand_financials_2.head()
```

```python
# Rename the column headers
cand_financials_2 = cand_financials_2.rename(columns={"Cand_Id" : "fec_candidate_id"})
cand_financials_2.head() 
```

```python
cand_financials_2 = cand_financials_2.drop(['Coverage_End_Date', 'Coverage_Start_Date'], axis=1)
cand_financials_2.head()
```

```python
connection_string = "root:AbuZaidSZK007@localhost:3306/election_prediction_db"
engine = create_engine(f'mysql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
# Clean the data by dropping duplicates and setting the index
cand_financials_2.drop_duplicates("fec_candidate_id", inplace=True)
```

```python
cand_financials_2.set_index("fec_candidate_id", inplace=True)
```

```python
cand_financials_2.to_sql(name='candidate_finance', con=engine, if_exists='append', index=True)
```
