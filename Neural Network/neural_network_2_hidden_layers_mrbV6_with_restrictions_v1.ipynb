```python
# Matthew Berning
# Dependencies
import numpy as np
import pandas as pd
import numpy
from numpy import array
from sqlalchemy import create_engine
import pymysql
```

```python
election_dataset_raw = pd.read_csv('../database/data_retrieval_clean.csv')
election_dataset_raw.head()
```

```python
election_dataset = election_dataset_raw.drop(['id','st_abbrev', 'district', "Industries_not_classified4", "state_dist"], axis=1)
```

```python
election_dataset_restricted = election_dataset[["party","Agriculture__forestry__fishing_and_hunting__and_mining", "100_000_to_149_999a", "American", "Less_than_50_000", "Public_transportation_excluding_taxicab", "ScotchIrish", "50_000_to_99_999", "Civilian_veterans", "Mining__quarrying__and_oil_and_gas_extraction2", "Unemployed", "White", "1_500_to_1_999b", "250_to_399", "500_to_999a", "Born_in_United_States", "Foreign_born", "Housing_units_without_a_mortgage", "Irish", "Median_dollars4", "Owneroccupied", "Asian", "Less_than_250", "Mining__quarrying__and_oil_and_gas_extraction3", "Native", "50_000_to_74_999", "Arab", "English", "Median_dollars2", "Median_dollars3", "Mining__quarrying__and_oil_and_gas_extraction1", "Scottish", "Prof__scientific__and_mngmnt__and_admin_and_waste_mngmnt_ser", "150_000_to_199_999", "400_to_599", "Abroad", "Born_in_PR__US_Is_areas_or_abroad_to_US_parents", "Graduate_or_professional_degree", "Median_dollars1", "Moved_in_2000_to_2009", "Occupied_units_paying_rent", "Same_state", "Some_other_race"]]

election_dataset_restricted.head()
```

```python
X = election_dataset_restricted.drop("party", axis=1)
y = election_dataset_restricted["party"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1, stratify=y)
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)


# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Create model and add layers
model = Sequential()
model.add(Dense(units=80, activation='relu', input_dim=42))
model.add(Dense(units=80, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=43,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:10])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels    : {list(y_test[:10])}")
```
